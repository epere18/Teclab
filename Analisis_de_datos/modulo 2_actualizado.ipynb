## Unidad 1 - Tema 3: Filtros de columnas y filas, integración de datos (join, append), transposición

```python
# Importamos de vuelta el archivo pero sin las últimas 3 filas
macro = pd.read_csv("datos_macro.csv", skipfooter = 3, engine = 'python')

# Filtramos las observaciones/filas que correspondan los países y variables que queremos
macro_f = macro[(macro["Country"].isin(["Argentina", "Chile", "Brazil"])) &
                  (macro["Subject Descriptor"] == "Gross domestic product, constant prices") &
                  (macro["Units"] == "Percent change") ]

macro_f
```

```python
# Eliminamos las columnas que ya no necesitamos
macro_f = macro_f.drop(columns = ["ï»¿WEO Country Code", "ISO", "WEO Subject Code",
                                  "Subject Descriptor", "Subject Notes",
                                  "Units", "Scale", "Country/Series-specific Notes", "Estimates Start After"])
macro_f
```

```python
# Transponemos la tabla para tenerla en formato largo, dejando fijo el país,
# convirtiendo los años de columnas a filas, y definiendo la variable de valores como crecimiento_pbi
macro_tr = pd.melt(macro_f, id_vars = ["Country"], var_name = "year", value_name = "crecimiento_pbi")
macro_tr
```

```python
macro_tr = macro_tr.rename(columns = {"Country": "pais"})
macro_tr
```

```python
# Ordenamos las filas, por país y por año
macro_tr = macro_tr.sort_values(by = ["pais", "year"])
macro_tr
```

```python
# Tipos de variables
macro_tr.dtypes
```

```python
tc_br.dtypes
```

```python
# Convertimos algunos de los campos a tipos de variables que nos sirvan: años como enteros, crecimiento como real
macro_tr = macro_tr.astype({"pais": str, "year": int, "crecimiento_pbi": float})
macro_tr.dtypes
```

```python
# Filtramos las observaciones anteriores a 2020, el último año con datos completos
macro_tr = macro_tr[macro_tr["year"] <= 2020]
macro_tr
```

```python
# Calculamos las principales estadísticas descriptivas de la variable crecimiento
macro_tr["crecimiento_pbi"].describe()
```

```python
tc_br
```

```python
# Concatenamos las tablas con el tipo de cambio de Brasil y de Argentina
tc = pd.concat([tc_br, tc_ar])

# Combinamos las dos tablas: la de tiene la serie del crecimiento del PBI y la del tipo de cambio
macro_tc = pd.merge(macro_tr, tc, left_on = ["pais", "year"], right_on = ["pais", "year"], how = 'outer')
macro_tc
```

## Unidad 1 - Tema 4. Agregaciones

```python
# Agregamos por país, usando el promedio (promedio de la tasa de crecimiento del PBI)
macro_ag = macro_tr.groupby(["pais"], as_index = False).agg({"crecimiento_pbi":"mean"})
macro_ag
```

### Calculá un campo nuevo, densidad (población/superficie) y usar la función descrita sobre ese campo nuevo.

```python
import pandas as pd
import numpy as np


poblacion_df = pd.read_csv("poblacion.csv", encoding = "latin-1 ")

poblacion_df
```

```python
hogares_viviendas_df = pd.read_csv("hogares_viviendas_superficie.csv", encoding = "latin-1 ")
hogares_viviendas_df
```

```python
# Filtrar y limpiar datos en el DataFrame de población
poblacion_df = poblacion_df[poblacion_df['provincia'] != 'Total País']
poblacion_df
```

```python
poblacion_df_densidad = pd.merge(poblacion_df, hogares_viviendas_df, left_on = ["provincia"], right_on = ["provincia"], how = 'left')
poblacion_df_densidad
```

```python
# Calcular la densidad poblacional
poblacion_df_densidad['densidad_poblacional'] = poblacion_df_densidad['poblacion_total'] / poblacion_df_densidad['superficie_km2']
poblacion_df_densidad
```

```python
# Guardar el DataFrame de población actualizado en un nuevo archivo CSV
poblacion_df_densidad.to_csv('poblacion_con_densidad.csv', index=False)
```
