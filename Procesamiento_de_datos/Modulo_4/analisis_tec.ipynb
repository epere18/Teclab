```python
import yfinance as yf

start_date = '2025-01-01'
end_date = '2025-07-24'

nvdia = yf.download('NVDA', start=start_date, end=end_date)

df = nvdia

df.head()
```

```python
# Calculamos la media movil de 30 dias
df['30_MA'] = df['Close'].rolling(window=30).mean()

# Calculamos la media movil de 90 dias
df['90_MA'] = df['Close'].rolling(window=90).mean()

# Graficamos el precio de cierre y las medias moviles
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Close'], label='NVDA Close Price', color='blue')
plt.plot(df.index, df['30_MA'], label='30-Day MA', color='orange')
plt.plot(df.index, df['90_MA'], label='90-Day MA', color='green')
plt.title('NVIDIA (NVDA) Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
print(df.columns)
df.columns = ['close', 'high', 'low', 'open', 'volume', '30_MA', '90_MA']
df.head()
```

```python
df.index = pd.DatetimeIndex(df.index)
```

```python
import pandas as pd
from finta import TA
import mplfinance as mpf

# Calcular Bandas de Bollinger con perÃ­odo 30
bbandas = TA.BBANDS(df, period=30)

# Unir las bandas superior e inferior en un DataFrame para graficarlas
bands_plot = pd.concat([bbandas.BB_UPPER, bbandas.BB_LOWER], axis=1)


# Crear el addplot para mplfinance
apd = mpf.make_addplot(bands_plot.tail(300))

# Graficar el precio con las Bandas de Bollinger
mpf.plot(df.tail(300), type='candle', style='charles', 
         title='NVDA Stock Price with Bollinger Bands', 
         ylabel='Price (USD)',
         addplot=apd, 
         volume=True, 
         figratio=(14, 7), figscale=1.2)


```
