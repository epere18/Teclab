# Manipulación de datos con Pandas

### Convierte un diccionario a una seria en Pandas

```python
import pandas as pd

diccionario_empleados  = {
    15379 : "hmaldonad@gmail.com",
    15380 : "mhurtado@gmail.com",
    15381 : "pperez@hotmail.com",
    15382 : "aperez@hotmail.com",
}

series_empleados = pd.Series(diccionario_empleados)
print(series_empleados)
```

### Crear series con etiquetas en Pandas

```python
lista_email = ["hmaldonad@gmail.com","mhurtado@gmail.com", "pperez@hotmail.com","aperez@hotmail.com"]
series_email = pd.Series(lista_email, index=[15379, 15380, 15381, 15382])
print(series_email)
```

### Ordenar series en Pandas

```python
serie_email_ordenada = series_empleados.sort_values(ascending=False)
print(serie_email_ordenada)
```

### Operaciones con series numericas

```python
lista_numerica = [30,20,50,90,80,70]
serie_numerica = pd.Series(lista_numerica)
print(serie_numerica)

#sumando 10 a cada elemento de la serie
serie_numerica_mas_10 = serie_numerica.add(10)
print("Sumando 10 a cada elemento de la serie:")
print(serie_numerica_mas_10)

#restando 10 a cada elemento de la serie
serie_numeriaca_menos_10 = serie_numerica.sub(10)
print("Restando 10 a cada elemento de la serie:")
print(serie_numeriaca_menos_10)

#multiplicando cada elemento de la serie por 10
serie_numerica_por_10 = serie_numerica.mul(10)
print("Multiplicando cada elemento de la serie por 10:")
print(serie_numerica_por_10)

#dividiendo cada elemento de la serie por 10
serie_numerica_dividido_10 = serie_numerica.div(10)
print("Dividiendo cada elemento de la serie por 10:")
print(serie_numerica_dividido_10)

#Calculando la media de la serie
media_serie_numerica = serie_numerica.mean()
print("Media de la serie numerica:")
print(media_serie_numerica)

#Calculando la desviación estándar de la serie
desviacion_serie_numerica = serie_numerica.std()
print("Desviación estándar de la serie numerica:")
print(desviacion_serie_numerica)

#Calculando la varianza de la serie
varianza_serie_numerica = serie_numerica.var()
print("Varianza de la serie numerica:")
print(varianza_serie_numerica)
```

### Operaciones con series de texto

```python
#Cambiando a minusculas
serie_minusculas_empleados = series_empleados.str.lower()
print("Cambiando a minusculas:")
print(serie_minusculas_empleados)

#Cambiando a mayusculas
serie_mayusculas_empleados = series_empleados.str.upper()
print("Cambiando a mayusculas:")
print(serie_mayusculas_empleados)

#Contando la cantidad de caracteres de cada elemento
serie_longitud_empleados = series_empleados.str.len()
print("Contando la cantidad de caracteres de cada elemento:")
print(serie_longitud_empleados)

#Reemplazando un dominio de correo
serie_reemplazo_dominio = series_empleados.str.replace("@gmail.com", "@outlook.com")
print("Reemplazando dominio de correo:")
print(serie_reemplazo_dominio)

#Contando la cantidad de veces que aparece un caracter
serie_contar_caracter = series_empleados.str.count("a")
print("Contando la cantidad de veces que aparece un caracter:")
print(serie_contar_caracter)
```

### Identificar valores faltantes en una serie en Pandas

```python
diccionario_nombre = {
    15379 : "Hector Maldonado",
    15738 : None,
    15380 : "Maria Hurtado",
    15381 : "Pedro Perez",
    15350 : None,
    15382 : "Ana Perez",
}

serie_nombre = pd.Series(diccionario_nombre)
print("Serie con valores faltantes:")
print(serie_nombre)

#Identificando valores faltantes
serie_valores_faltantes = serie_nombre.isna()
print("Identificando valores faltantes:")
print(serie_valores_faltantes)

#Contando valores faltantes
conteo_valores_faltantes = serie_nombre.isna().sum()
print("Contando valores faltantes:")
print(conteo_valores_faltantes)
```

### Funciones de agregación en serie de Pandas

```python
salarios_anuales = [1024000,1100000,985000,1600000]
serie_salarios = pd.Series(salarios_anuales)
print("Serie de salarios anuales:")
print(serie_salarios)

#Calculando la media de los salarios
media_salarios = serie_salarios.mean()
print("Media de los salarios anuales:")
print(media_salarios)

#Calculando la mediana de los salarios
mediana_salarios = serie_salarios.median()
print("Mediana de los salarios anuales:")
print(mediana_salarios)

#Calculando el valor máximo de los salarios
maximo_salarios = serie_salarios.max()
print("Valor máximo de los salarios anuales:")
print(maximo_salarios)

#Calculando el valor mínimo de los salarios
minimo_salarios = serie_salarios.min()
print("Valor mínimo de los salarios anuales:")
print(minimo_salarios)

#Calculando la desviación estándar de los salarios
desviacion_salarios = serie_salarios.std()
print("Desviación estándar de los salarios anuales:")
print(desviacion_salarios)

#Calculando la varianza de los salarios
varianza_salarios = serie_salarios.var()
print("Varianza de los salarios anuales:")
print(varianza_salarios)
```

### Unir dos dataFrames por un rango de fechas

```python
import pandas as pd

df_caja1 = pd.DataFrame({
    "fecha" : pd.date_range(start='2023-01-01', periods=5, freq='D'),
    "ventas_caja1" : [100, 200, 150, 300, 250]
})

df_caja2 = pd.DataFrame({
    "fecha" : pd.date_range(start='2023-01-01', periods=5, freq='D'),
    "ventas_caja2" : [120, 220, 180, 320, 270]
})

df_ventas_totales = pd.merge(df_caja1, df_caja2, on="fecha", how="outer")
df_ventas_totales["ventas_totales"] = df_ventas_totales[["ventas_caja1", "ventas_caja2"]].sum(axis=1)
print("DataFrame de ventas totales por fecha:")
print(df_ventas_totales)

```

### Convertir un archivo XLSX en un dataFrame en Pandas

```python
import pandas as pd

archivo = "Facturacion2023.xlsx"
df_facturacion = pd.read_excel(archivo)
print("DataFrame cargado desde el archivo XLSX:")
df_facturacion
```

```python
num_filas, num_columnas = df_facturacion.shape
print(f"El DataFrame tiene {num_filas} filas y {num_columnas} columnas.")

valores_faltantes = df_facturacion.isna().sum()
print("Valores faltantes en cada columna:")
print(valores_faltantes)
```

```python
df_facturacion["Provincia"] = df_facturacion["Provincia"].fillna("Desconocida")
print("Valores faltantes en la columna 'Provincia' han sido reemplazados por 'Desconocida'.")

df_facturacion["LOCALIDAD"] = df_facturacion["LOCALIDAD"].fillna("Desconocida")
print("Valores faltantes en la columna 'LOCALIDAD' han sido reemplazados por 'Desconocida'.")
```

### Reestructurar un dataFrame en Pandas

```python
import pandas as pd

# Opcional: formato de visualización de floats

df_restructurado = df_facturacion.pivot_table(index='Sector', columns='Sector_OR', values='Neto_Gravado', aggfunc='sum').fillna(0)

# Calcular la fila total (suma por columnas)
fila_total = df_restructurado.sum().to_frame().T

# Agregarle un índice que diga "TOTAL"
fila_total.index = ['TOTAL']

# Unir el total al final del DataFrame
df_restructurado = pd.concat([df_restructurado, fila_total])

# Aplicar formato de moneda en pesos argentinos
df_restructurado1 = df_restructurado.applymap(lambda x: f"${x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))

df_restructurado1.head(10)  # Mostrar las primeras 10 filas del DataFrame reestructurado
```

### Manejo de agrupamiento sobre dataframes en Pandas

```python
df_facturacion["Ganancia_REP"] = df_facturacion["REP_s_Iva"] - df_facturacion["REP_Costo_s_Iva"]

sector_agrupado = df_facturacion.groupby("Sector")
resultados_agrupados = sector_agrupado.agg({
    'Neto_Gravado': 'sum',
    'REP_s_Iva': 'sum',
    'REP_Costo_s_Iva': 'sum',
    'Ganancia_REP': 'sum'
}).reset_index()
print("Resultados agrupados por sector:")
resultados_agrupados
```
