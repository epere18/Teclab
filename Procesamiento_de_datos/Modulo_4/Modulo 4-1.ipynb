# Estilos de configuraciÃ³n

La funciÃ³n plt.style en Matplotlib te permite seleccionar estilos estÃ©ticos predefinidos para tus grÃ¡ficos. AdemÃ¡s de los estilos predefinidos, tambiÃ©n puedes crear tus propios estilos personalizados si lo deseas.

Algunos de los estilos predefinidos que puedes usar con plt.style incluyen:

'ggplot': Este estilo estÃ¡ inspirado en el estilo de los grÃ¡ficos en el software de visualizaciÃ³n de datos R, ggplot2. Tiene colores suaves y lÃ­neas mÃ¡s gruesas.

'seaborn': Este estilo se basa en el paquete Seaborn y proporciona grÃ¡ficos limpios y modernos con colores agradables.

'fivethirtyeight': Este estilo estÃ¡ inspirado en el sitio web FiveThirtyEight, conocido por su cobertura de estadÃ­sticas y polÃ­tica. Tiene un estilo distintivo y colorido.

'bmh': Este estilo sigue el estilo de la biblioteca de grÃ¡ficos estadÃ­sticos de MATLAB (Matlab Plotting Toolbox).

'dark_background': Este estilo utiliza un fondo oscuro y colores claros para los elementos grÃ¡ficos.

# VisualizaciÃ³n de los grÃ¡ficos

```python
# Crear un rango de valores para x (por ejemplo, de 0 a 2*pi)
x = np.linspace(0, 2 * np.pi, 100)  # 100 puntos entre 0 y 2*pi

# Crear una nueva figura
plt.figure(figsize=(8, 6))

# Calcular los valores de seno y coseno para cada valor de x
y_sin = np.sin(x)
y_cos = np.cos(x)

# Graficar la funciÃ³n seno
plt.plot(x, y_sin, label='Seno', color='blue')

# Graficar la funciÃ³n coseno
plt.plot(x, y_cos, label='Coseno', color='red')

# Agregar leyenda
plt.legend()

# Agregar tÃ­tulo y etiquetas de los ejes
plt.title('Funciones Seno y Coseno')
plt.xlabel('x')
plt.ylabel('y')

# Mostrar la grÃ¡fica
plt.grid(True)
plt.show()
```

# Grafico de lineas simples

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
# Datos para el eje x (tiempo)
tiempo = [0, 1, 2, 3, 4, 5]

# Datos para el eje y (valores)
valores = [0, 1, 4, 9, 16, 25]

# Crear el grÃ¡fico de lÃ­neas
plt.plot(tiempo, valores)

# Etiquetas y tÃ­tulo
plt.xlabel('Tiempo')
plt.ylabel('Valores')
plt.title('GrÃ¡fico de LÃ­neas')

# Mostrar el grÃ¡fico
plt.grid(True)  # Mostrar la cuadrÃ­cula
plt.show()
```

# Estilo de lineas

```python
# Datos para el eje x (tiempo)
tiempo = [0, 1, 2, 3, 4, 5]

# Datos para el eje y (valores)
valores = [0, 1, 4, 9, 16, 25]

# Crear el grÃ¡fico de lÃ­neas con diferentes estilos
plt.plot(tiempo, valores, linestyle='-', label='Solido')  # LÃ­nea sÃ³lida
plt.plot(tiempo, [v + 2 for v in valores], linestyle='--', label='Discontinua')  # LÃ­nea discontinua
plt.plot(tiempo, [v - 2 for v in valores], linestyle='-.', label='Punteado y discontinua')  # LÃ­nea punteada y discontinua
plt.plot(tiempo, [v * 2 for v in valores], linestyle=':', label='Punteado')  # LÃ­nea punteada

# Etiquetas y tÃ­tulo
plt.xlabel('Tiempo')
plt.ylabel('Valores')
plt.title('GrÃ¡fico de LÃ­neas con Diferentes Estilos')

# Agregar una leyenda
plt.legend()

# Mostrar el grÃ¡fico
plt.grid(True)  # Mostrar la cuadrÃ­cula
plt.show()
```

## GrÃ¡fico de barras discreto

```python
# Datos para las barras (por ejemplo, categorÃ­as)
categorias = ['A', 'B', 'C', 'D', 'E']

# Datos para las alturas de las barras (por ejemplo, frecuencias)
alturas = [10, 15, 20, 25, 30]

# Crear el grÃ¡fico de barras
plt.bar(categorias, alturas)

# Etiquetas y tÃ­tulo
plt.xlabel('CategorÃ­as')
plt.ylabel('Alturas')
plt.title('GrÃ¡fico de Barras Discreto')

# Mostrar el grÃ¡fico
plt.grid(True)  # Mostrar la cuadrÃ­cula
plt.show()
```

```python
# Datos para las barras (por ejemplo, categorÃ­as)
categorias = ['A', 'B', 'C', 'D', 'E']

# Datos para las alturas de las barras (por ejemplo, frecuencias)
alturas = [10, 15, 20, 25, 30]

# Crear el grÃ¡fico de barras
plt.barh(categorias, alturas, color='lightgreen')

# Etiquetas y tÃ­tulo
plt.xlabel('CategorÃ­as')
plt.ylabel('Alturas')
plt.title('GrÃ¡fico de Barras Discreto')

# Mostrar el grÃ¡fico
plt.grid(True)  # Mostrar la cuadrÃ­cula
plt.show()
```

```python
# grafico de barras agrupadas

import numpy as np
import matplotlib.pyplot as plt

# Datos para las barras (por ejemplo, categorÃ­as)
categorias = ['A', 'B', 'C', 'D', 'E']

# Datos para las alturas de las barras para el primer grupo
alturas_grupo1 = [10, 15, 20, 25, 30]

# Datos para las alturas de las barras para el segundo grupo
alturas_grupo2 = [15, 20, 25, 30, 35]

# Definir el ancho de las barras
ancho_barra = 0.35

# Calcular la posiciÃ³n de los grupos
posicion_grupo1 = np.arange(len(categorias))
posicion_grupo2 = [x + ancho_barra for x in posicion_grupo1]

# Crear el grÃ¡fico de barras agrupadas
plt.bar(posicion_grupo1, alturas_grupo1, width=ancho_barra, label='Grupo 1')
plt.bar(posicion_grupo2, alturas_grupo2, width=ancho_barra, label='Grupo 2')

# Etiquetas y tÃ­tulo
plt.xlabel('CategorÃ­as')
plt.ylabel('Alturas')
plt.title('GrÃ¡fico de Barras Agrupadas')
plt.xticks([x + ancho_barra / 2 for x in posicion_grupo1], categorias)

# Agregar leyenda
plt.legend()

# Mostrar el grÃ¡fico
plt.grid(True)  # Mostrar la cuadrÃ­cula
plt.show()
```

```python
# Grafico de barras apiladas

import numpy as np
import matplotlib.pyplot as plt

# Datos para las categorÃ­as
categorias = ['A', 'B', 'C', 'D', 'E']

# Datos para las alturas de las barras para el primer conjunto
alturas_conjunto1 = [10, 15, 20, 25, 30]

# Datos para las alturas de las barras para el segundo conjunto
alturas_conjunto2 = [15, 20, 25, 30, 35]

# PosiciÃ³n de las barras
posicion = np.arange(len(categorias))

# Crear el grÃ¡fico de barras apiladas
plt.bar(posicion, alturas_conjunto1, label='Conjunto 1')
plt.bar(posicion, alturas_conjunto2, bottom=alturas_conjunto1, label='Conjunto 2')

# Etiquetas y tÃ­tulo
plt.xlabel('CategorÃ­as')
plt.ylabel('Alturas')
plt.title('GrÃ¡fico de Barras Apiladas')

# Agregar etiquetas de categorÃ­as en el eje x
plt.xticks(posicion, categorias)

# Agregar leyenda
plt.legend()

# Mostrar el grÃ¡fico
plt.grid(True)  # Mostrar la cuadrÃ­cula
plt.show()
```

## Histograma

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

# Generar datos aleatorios con una distribuciÃ³n normal
datos = np.random.randn(1000)

# Crear el histograma
plt.hist(datos, bins=30, color='blue', edgecolor='black')

# bins especifica el nÃºmero de intervalos en los que se dividirÃ¡n los datos.

# Etiquetas y tÃ­tulo
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma')

# Mostrar el histograma
plt.grid(True)  # Mostrar la cuadrÃ­cula
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios con diferentes distribuciones
x1= np.random.normal(loc=0, scale=1, size=1000),
x2=np.random.normal(loc=2, scale=1.5, size=1000),
x3=np.random.normal(loc=-2, scale=0.5, size=1000)

kwargs=dict(histtype='stepfilled', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

# Diagrama de caja

```python
# Generar datos aleatorios para cuatro grupos
datos = np.random.normal(0, 0.7, 100)

# Crear el diagrama de caja
plt.boxplot(datos)

# Etiquetas y tÃ­tulo
plt.xlabel('Grupo')
plt.ylabel('Valor')
plt.title('Diagrama de Caja')

# Mostrar el diagrama de caja
plt.grid(True)  # Mostrar la cuadrÃ­cula
plt.show()
```

```python
#AquÃ­ te muestro un ejemplo utilizando Matplotlib para crear un grÃ¡fico de caja y bigotes con tres variables distintas en el mismo grÃ¡fico:

# Generar datos aleatorios para tres grupos
datos = np.random.normal(0, 1, 100)

# Agregar valores atÃ­picos a los datos del primer grupo
datos = np.concatenate([datos, [5, 4.3, -5]])

plt.figure(figsize=(8, 6))

# Crear el diagrama de caja con los tres conjuntos de datos
plt.boxplot(datos)

# Etiquetas y tÃ­tulo
plt.xlabel('Grupo')
plt.ylabel('Valor')
plt.title('Diagrama de Caja con Valores AtÃ­picos')

# Mostrar el diagrama de caja
plt.grid(True)  # Mostrar la cuadrÃ­cula
plt.show()
```

### Ejemplo utilizando Matplotlib para crear un grÃ¡fico de caja y bigotes con tres variables distintas en el mismo grÃ¡fico:

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=2, scale=1.5, size=100)
data3 = np.random.normal(loc=-2, scale=0.5, size=100)

# Crear una lista de datos
data = [data1, data2, data3]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Crear el grÃ¡fico de caja y bigotes
ax.boxplot(data)

# AÃ±adir etiquetas a los ejes
ax.set_xticklabels(['Variable 1', 'Variable 2', 'Variable 3'])

# AÃ±adir tÃ­tulo y etiquetas
ax.set_title('GrÃ¡fico de Caja y Bigotes de Tres Variables')
ax.set_ylabel('Valor')

# Mostrar el grÃ¡fico
plt.show()
```
