# Estilos de configuración

La función plt.style en Matplotlib te permite seleccionar estilos estéticos predefinidos para tus gráficos. Además de los estilos predefinidos, también puedes crear tus propios estilos personalizados si lo deseas.

Algunos de los estilos predefinidos que puedes usar con plt.style incluyen:

'ggplot': Este estilo está inspirado en el estilo de los gráficos en el software de visualización de datos R, ggplot2. Tiene colores suaves y líneas más gruesas.

'seaborn': Este estilo se basa en el paquete Seaborn y proporciona gráficos limpios y modernos con colores agradables.

'fivethirtyeight': Este estilo está inspirado en el sitio web FiveThirtyEight, conocido por su cobertura de estadísticas y política. Tiene un estilo distintivo y colorido.

'bmh': Este estilo sigue el estilo de la biblioteca de gráficos estadísticos de MATLAB (Matlab Plotting Toolbox).

'dark_background': Este estilo utiliza un fondo oscuro y colores claros para los elementos gráficos.

# Visualización de los gráficos

```python
# Crear un rango de valores para x (por ejemplo, de 0 a 2*pi)
x = np.linspace(0, 2 * np.pi, 100)  # 100 puntos entre 0 y 2*pi

# Crear una nueva figura
plt.figure(figsize=(8, 6))

# Calcular los valores de seno y coseno para cada valor de x
y_sin = np.sin(x)
y_cos = np.cos(x)

# Graficar la función seno
plt.plot(x, y_sin, label='Seno', color='blue')

# Graficar la función coseno
plt.plot(x, y_cos, label='Coseno', color='red')

# Agregar leyenda
plt.legend()

# Agregar título y etiquetas de los ejes
plt.title('Funciones Seno y Coseno')
plt.xlabel('x')
plt.ylabel('y')

# Mostrar la gráfica
plt.grid(True)
plt.show()
```

# Grafico de lineas simples

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
# Datos para el eje x (tiempo)
tiempo = [0, 1, 2, 3, 4, 5]

# Datos para el eje y (valores)
valores = [0, 1, 4, 9, 16, 25]

# Crear el gráfico de líneas
plt.plot(tiempo, valores)

# Etiquetas y título
plt.xlabel('Tiempo')
plt.ylabel('Valores')
plt.title('Gráfico de Líneas')

# Mostrar el gráfico
plt.grid(True)  # Mostrar la cuadrícula
plt.show()
```

# Estilo de lineas

```python
# Datos para el eje x (tiempo)
tiempo = [0, 1, 2, 3, 4, 5]

# Datos para el eje y (valores)
valores = [0, 1, 4, 9, 16, 25]

# Crear el gráfico de líneas con diferentes estilos
plt.plot(tiempo, valores, linestyle='-', label='Solido')  # Línea sólida
plt.plot(tiempo, [v + 2 for v in valores], linestyle='--', label='Discontinua')  # Línea discontinua
plt.plot(tiempo, [v - 2 for v in valores], linestyle='-.', label='Punteado y discontinua')  # Línea punteada y discontinua
plt.plot(tiempo, [v * 2 for v in valores], linestyle=':', label='Punteado')  # Línea punteada

# Etiquetas y título
plt.xlabel('Tiempo')
plt.ylabel('Valores')
plt.title('Gráfico de Líneas con Diferentes Estilos')

# Agregar una leyenda
plt.legend()

# Mostrar el gráfico
plt.grid(True)  # Mostrar la cuadrícula
plt.show()
```

## Gráfico de barras discreto

```python
# Datos para las barras (por ejemplo, categorías)
categorias = ['A', 'B', 'C', 'D', 'E']

# Datos para las alturas de las barras (por ejemplo, frecuencias)
alturas = [10, 15, 20, 25, 30]

# Crear el gráfico de barras
plt.bar(categorias, alturas)

# Etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Alturas')
plt.title('Gráfico de Barras Discreto')

# Mostrar el gráfico
plt.grid(True)  # Mostrar la cuadrícula
plt.show()
```

```python
# Datos para las barras (por ejemplo, categorías)
categorias = ['A', 'B', 'C', 'D', 'E']

# Datos para las alturas de las barras (por ejemplo, frecuencias)
alturas = [10, 15, 20, 25, 30]

# Crear el gráfico de barras
plt.barh(categorias, alturas, color='lightgreen')

# Etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Alturas')
plt.title('Gráfico de Barras Discreto')

# Mostrar el gráfico
plt.grid(True)  # Mostrar la cuadrícula
plt.show()
```

```python
# grafico de barras agrupadas

import numpy as np
import matplotlib.pyplot as plt

# Datos para las barras (por ejemplo, categorías)
categorias = ['A', 'B', 'C', 'D', 'E']

# Datos para las alturas de las barras para el primer grupo
alturas_grupo1 = [10, 15, 20, 25, 30]

# Datos para las alturas de las barras para el segundo grupo
alturas_grupo2 = [15, 20, 25, 30, 35]

# Definir el ancho de las barras
ancho_barra = 0.35

# Calcular la posición de los grupos
posicion_grupo1 = np.arange(len(categorias))
posicion_grupo2 = [x + ancho_barra for x in posicion_grupo1]

# Crear el gráfico de barras agrupadas
plt.bar(posicion_grupo1, alturas_grupo1, width=ancho_barra, label='Grupo 1')
plt.bar(posicion_grupo2, alturas_grupo2, width=ancho_barra, label='Grupo 2')

# Etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Alturas')
plt.title('Gráfico de Barras Agrupadas')
plt.xticks([x + ancho_barra / 2 for x in posicion_grupo1], categorias)

# Agregar leyenda
plt.legend()

# Mostrar el gráfico
plt.grid(True)  # Mostrar la cuadrícula
plt.show()
```

```python
# Grafico de barras apiladas

import numpy as np
import matplotlib.pyplot as plt

# Datos para las categorías
categorias = ['A', 'B', 'C', 'D', 'E']

# Datos para las alturas de las barras para el primer conjunto
alturas_conjunto1 = [10, 15, 20, 25, 30]

# Datos para las alturas de las barras para el segundo conjunto
alturas_conjunto2 = [15, 20, 25, 30, 35]

# Posición de las barras
posicion = np.arange(len(categorias))

# Crear el gráfico de barras apiladas
plt.bar(posicion, alturas_conjunto1, label='Conjunto 1')
plt.bar(posicion, alturas_conjunto2, bottom=alturas_conjunto1, label='Conjunto 2')

# Etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Alturas')
plt.title('Gráfico de Barras Apiladas')

# Agregar etiquetas de categorías en el eje x
plt.xticks(posicion, categorias)

# Agregar leyenda
plt.legend()

# Mostrar el gráfico
plt.grid(True)  # Mostrar la cuadrícula
plt.show()
```

## Histograma

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')

# Generar datos aleatorios con una distribución normal
datos = np.random.randn(1000)

# Crear el histograma
plt.hist(datos, bins=30, color='blue', edgecolor='black')

# bins especifica el número de intervalos en los que se dividirán los datos.

# Etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Histograma')

# Mostrar el histograma
plt.grid(True)  # Mostrar la cuadrícula
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos aleatorios con diferentes distribuciones
x1= np.random.normal(loc=0, scale=1, size=1000),
x2=np.random.normal(loc=2, scale=1.5, size=1000),
x3=np.random.normal(loc=-2, scale=0.5, size=1000)

kwargs=dict(histtype='stepfilled', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

# Diagrama de caja

```python
# Generar datos aleatorios para cuatro grupos
datos = np.random.normal(0, 0.7, 100)

# Crear el diagrama de caja
plt.boxplot(datos)

# Etiquetas y título
plt.xlabel('Grupo')
plt.ylabel('Valor')
plt.title('Diagrama de Caja')

# Mostrar el diagrama de caja
plt.grid(True)  # Mostrar la cuadrícula
plt.show()
```

```python
#Aquí te muestro un ejemplo utilizando Matplotlib para crear un gráfico de caja y bigotes con tres variables distintas en el mismo gráfico:

# Generar datos aleatorios para tres grupos
datos = np.random.normal(0, 1, 100)

# Agregar valores atípicos a los datos del primer grupo
datos = np.concatenate([datos, [5, 4.3, -5]])

plt.figure(figsize=(8, 6))

# Crear el diagrama de caja con los tres conjuntos de datos
plt.boxplot(datos)

# Etiquetas y título
plt.xlabel('Grupo')
plt.ylabel('Valor')
plt.title('Diagrama de Caja con Valores Atípicos')

# Mostrar el diagrama de caja
plt.grid(True)  # Mostrar la cuadrícula
plt.show()
```

### Ejemplo utilizando Matplotlib para crear un gráfico de caja y bigotes con tres variables distintas en el mismo gráfico:

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=2, scale=1.5, size=100)
data3 = np.random.normal(loc=-2, scale=0.5, size=100)

# Crear una lista de datos
data = [data1, data2, data3]

# Crear la figura y los ejes
fig, ax = plt.subplots()

# Crear el gráfico de caja y bigotes
ax.boxplot(data)

# Añadir etiquetas a los ejes
ax.set_xticklabels(['Variable 1', 'Variable 2', 'Variable 3'])

# Añadir título y etiquetas
ax.set_title('Gráfico de Caja y Bigotes de Tres Variables')
ax.set_ylabel('Valor')

# Mostrar el gráfico
plt.show()
```
